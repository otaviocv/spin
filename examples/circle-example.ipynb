# Circle Example

```python
import numpy as np
from spin.distances import l2_distance_matrix
from spin import NeighborhoodSPIN, SideToSideSPIN
import matplotlib.pyplot as plt
%matplotlib inline
```

## Generating Data

```python
n_points = 1000
```

```python
r_coords = np.random.normal(1, 0.1, size=n_points)
t_coords = np.random.uniform(0, 2*np.pi, size=n_points)
```

```python
x_coords = r_coords * np.cos(t_coords)
y_coords = r_coords * np.sin(t_coords)
```

### Visualizing data

```python
f, ax = plt.subplots(figsize=(5, 5), dpi=100)
ax.scatter(x_coords, y_coords, s=.8)
```

```python
points = np.array([x_coords, y_coords])
```

### Distance Matrix

```python
distances = l2_distance_matrix(points, points)
```

```python
f, ax = plt.subplots(figsize=(5, 5), dpi=100)
ax.matshow(distances, cmap="plasma")
```
